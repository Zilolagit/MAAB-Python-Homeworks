```python
import pandas as pd
import numpy as np

df = pd.read_csv('./data/flights.csv')
df
```

```python
df.groupby('AIRLINE').agg({'DAY': 'mean', 'MONTH': 'median'})
```

```python
flights = df.groupby(['ORG_AIR', 'DEST_AIR']).agg(
    sum_cancelled = pd.NamedAgg(column='CANCELLED', aggfunc='sum'),
    mean_cancelled = pd.NamedAgg(column='CANCELLED', aggfunc='mean'),
)
flights
```

```python

df.groupby(['MONTH', 'WEEKDAY']).agg({'AIR_TIME': ['sum', 'mean'], 'DIVERTED': 'sum', 'CANCELLED': 'sum' } ).astype('int').reset_index()
```

```python
flights = df.groupby(['MONTH', 'WEEKDAY']).agg(
    sum_air_time = pd.NamedAgg(column='AIR_TIME', aggfunc='sum'),
    mean_air_time = pd.NamedAgg(column='AIR_TIME', aggfunc='mean'),
    sum_diverted = pd.NamedAgg(column='DIVERTED', aggfunc='sum'),
    sum_cancelled = pd.NamedAgg(column='CANCELLED', aggfunc='sum'),
).astype('int').reset_index()
flights
```
