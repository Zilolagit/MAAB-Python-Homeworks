```python
# N=1
import sqlite3
import pandas as pd

conn = sqlite3.connect('./task/population.db')

population_df = pd.read_sql_query("Select * from population", conn)

conn.close()

population_df
```

```python
# N=2
salary_analysis_df  = pd.read_excel('./task/population_salary_analysis.xlsx')
salary_analysis_df
```

```python
salary_categories = {
    "till $200,000": (0, 200000),
    "$200,001 - $400,000": (200001, 400000),
    "$400,001 - $600,000": (400001, 600000),
    "$600,001 - $800,000": (600001, 800000),
    "$800,001 - $1,000,000": (800001, 1000000),
    "$1,000,001 - $1,200,000": (1000001, 1200000),
    "$1,200,001 and above": (1200001, float('inf'))
}
def custom_salary(salary):
    for i, (begin, start) in salary_categories.items():
        if salary >= begin and salary <= start:
            return i
    return None

population_df['SalaryCategory'] = population_df['salary'].apply(custom_salary)
population_df
```

```python
# Population Number
total_population = len(population_df)
total_population
salary_distribution = population_df['SalaryCategory'].value_counts()
salary_distribution
salary_distribution = salary_distribution.reset_index()
salary_distribution
```

```python
# Percentage
salary_distribution['Percentage'] = (salary_distribution['count'] / total_population) * 100
salary_distribution
```

```python
salary_stats = population_df.groupby('SalaryCategory')['salary'].agg(['mean', 'median'])
salary_stats
```

```python
salary_stats = population_df.groupby('state').agg({'salary': ['mean', 'median']}).reset_index()
salary_stats
```
